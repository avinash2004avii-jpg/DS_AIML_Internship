```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"C:\Users\achar\DS_AI_internship\src\day27_Logistics_regression\insurance_data.csv")
df.head()
```

```python
df.shape
```

```python
plt.scatter(df.age,df.bought_insurance,color ='red',marker ='*')
```

```python
x = df[['age']]
y = df['bought_insurance']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=10)
```

```python
x_test
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
coef = model.coef_
coef
```

```python
intercept = model.intercept_
intercept
```

```python
y_predicted = model.predict(x_test)
```

```python
y_predicted
```

```python
df['probs'] = model.predict_proba(df[['age']])[:,1]
df
```

```python
plt.scatter(df['age'],df['bought_insurance'],color = 'red',marker = '*')
plt.plot(df['age'],df['probs'],color = 'blue')
```

```python
import math
def sigmoid(z):
    return 1/ (1 + math.exp(-z))
```

```python
def predicting_age(age):
    z = coef * age + (intercept)
    y = sigmoid(z)
    return y
```

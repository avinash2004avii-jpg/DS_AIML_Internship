```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
data = {
'Weather':['Sunny','Sunny','Cloudy','Rain','Rain','Rain','Cloudy','Sunny','Sunny','Rain','Sunny','Cloudy','Cloudy','Rain'],

'Temperature':['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild'],

'Humidity':['High','High','High','High','Normal','Normal','Normal','High','Normal','Normal','Normal','High','Normal','High'],

'Wind':['Weak','Strong','Weak','Weak','Weak','Strong','Strong','Weak','Weak','Weak','Strong','Strong','Weak','Strong'],

'PlayFootball':['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Display dataset
df
```

```python
df.head()
```

```python
le=LabelEncoder()

for col in df.columns:
    df[col]=le.fit_transform(df[col])
df    
```

```python
x= df.drop('PlayFootball',axis=1)
y=df['PlayFootball']
print("Features(x):")
display(x)

print("target(y)")
display(y)
      
```

```python
model=DecisionTreeClassifier()
model.fit(x,y)
```

```python
print("Model trained successfully")
```

```python
sample=[[2,0,0,1]]
prediction=model.predict(sample)
prediction
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(10,10))
plot_tree(
    model,
    feature_names=x.columns,
    class_names=['No','Yes'],
    filled=True,
    rounded=True,
    fontsize=12,
)

plt.title("Decision Tree using ID3 Algorithm")
plt.show()
```

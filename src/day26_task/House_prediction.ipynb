```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\achar\OneDrive\Desktop\homeprices.csv")
df.head()
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area,df.price,color='black',marker='*')
```

```python
area = df.drop('price',axis = 'columns')
area
```

```python
price = df.price
price
```

```python
#equation for linear regression y= mx + b
reg = linear_model.LinearRegression()
reg.fit(area,price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#To predict the new area 3500
new_data = pd.DataFrame([[3500]],columns=['area'])
prediction = reg.predict(new_data)
prediction = np.round(prediction,2)
prediction
```

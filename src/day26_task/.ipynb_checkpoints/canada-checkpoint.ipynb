```python
import pandas as pd 
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\achar\DS_AI_internship\src\day26_task\canada_per_capita_income.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
plt.xlabel('year')
plt.ylabel('per_capita_income')
plt.scatter(df['year'], df['per capita income (US$)'], color='black', marker='*')
```

```python
plt.plot(df['year'], df['per capita income (US$)'], marker='*')
plt.xlabel('Year')
plt.ylabel('per capita income (US$)')
plt.title('Per Capita Income Over Years')
plt.show()
```

```python
per_capita_income = df['per capita income (US$)']
per_capita_income
```

```python
X = df[['year']]   
y = df['per capita income (US$)']
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
new_data=pd.DataFrame({'year' : [2017,2018,2019,2020]})
prediction=reg.predict(new_data)
prediction=np.round(prediction,2)
prediction
```

```python
new_data['per capita income (US$)'] = prediction
```

```python
new_data.to_csv('canada_per_capita_income.csv',index = False,header=False,mode="a")
```
